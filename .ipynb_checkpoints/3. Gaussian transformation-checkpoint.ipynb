```python
from sklearn.datasets import fetch_openml

boston = fetch_openml(name='boston', version=1, as_frame=True)
df = boston.frame[['INDUS', 'RM', 'AGE',	'DIS', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']]

df.head()
```

```python
from scipy.stats import skew

def plot_data(col, df=df, target='MEDV'):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.hist(df[col], bins=10, color='yellow', edgecolor='black')
    plt.title(f"Distribution of {col}", fontsize=14)
    plt.xlabel('Values', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(True)
    skewness_value = skew(df[col])
    plt.text(x=0.95, y=0.95, s=f"Skewness: {skewness_value:.3f}", transform=plt.gca().transAxes, fontsize=12, color='darkred', 
             horizontalalignment='right', verticalalignment='top', bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.3'))
    
    plt.subplot(1, 2, 2)
    plt.scatter(df[col], df[target], color='seagreen', alpha=0.7, edgecolor='k', s=40)
    plt.title(f'Feature: {col}', fontsize=14)
    plt.xlabel(f'{col}', fontsize=12)
    plt.ylabel('Target (MEDV)', fontsize=12)
    plt.grid(True)
    plt.show()
```

```python
for col in df.columns[:-1]:
    plot_data(col)
```

# Feature Transformation Decision Guide

Below is a quick reference for choosing the right Gaussian‑style transformation based on a feature’s skewness, outlier profile, and domain.

| Case                                | Skewness & Outliers                     | Transform(s)                            | Notes                                                                 |
|-------------------------------------|-----------------------------------------|-----------------------------------------|-----------------------------------------------------------------------|
| **A. Slight skew**                  | –0.5 ≤ γ ≤ 0.5, few outliers             | None or **√**                                | Often no transform needed; √ can gently symmetrize if desired.        |
| **B. Moderate right skew**          | 0.5 < γ ≤ 1, moderate skew, few outliers | **Log** or **Box‑Cox**                          | Log smooths more; Box‑Cox can optimize λ for best normality.         |
| **C. Moderate left skew**           | –1 ≤ γ < –0.5, moderate left skew        | **Exponential** (power > 1)                 | E.g. x² or exp(x) to stretch upper tail.                             |
| **D. Strong right skew**            | γ > 1, heavy skew, few outliers         | **Reciprocal** (1/x) or Log                 | Reciprocal flips + compresses; log if you want monotonic but less flip. |
| **E. Strong left skew**             | γ < –1, heavy left skew                 | **Exponential** (eˣ)                        | Use a power > 1 to push small values downward.                       |
| **F. Many outliers (any skew)**     | γ any, many outliers                    | **Quantile Transformer** or **Yeo‑Johnson**     | Quantile → exact Gaussian shape; YJ → robust power transform with λ.  |
| **G. Mixed sign values**            | γ any, includes negatives/zeros         | **Yeo‑Johnson**                              | Box‑Cox fails at ≤ 0; YJ handles full real line.                      |
| **H. Low/medium correlation**       | any skew                                | **Quantile** or **Box‑Cox** / **Yeo‑Johnson**       | Rank‑based (Quantile) can boost weak linearity vs. target.            |


## Detailed Rationales

### √ (Square Root)
- **Use**: small positive skew (γ between 0.3 and 0.7), count data (e.g., number of clicks).  
- **Effect**: mild compression of the right tail.

---

### Logarithm
- **Use**: moderate right skew (γ between 0.7 and 1.5), strictly positive features.  
- **Effect**: strong compression of large values; stabilizes variance.

---

### Reciprocal (1/x)
- **Use**: very heavy right skew (γ > 1.5), no zeros in the data.  
- **Effect**: flips the axis and compresses large values most aggressively.

---

### Exponential / Power (>1)
- **Use**: left-skewed features (γ < –0.5) to stretch the upper tail.  
- **Effect**: accentuates differences among larger values.

---

### Box-Cox
- **Use**: strictly positive data with moderate to heavy skew.  
- **Effect**: finds an optimal λ parameter to best approximate a Gaussian distribution.

---

### Yeo-Johnson
- **Use**: data that may include zeros or negatives, with moderate to heavy skew.  
- **Effect**: like Box-Cox but supports the full real line (≤ 0 included).

---

### Quantile Transformer
- **Use**: any skew or outlier situation, especially if you need an **exact** normal distribution or have **many** outliers.  
- **Effect**: maps feature quantiles to a standard normal distribution (note: destroys original scale and ordering).


### Log Transformation

**Goal:**  Compress moderate-to-large positive values to reduce right‑skew and stabilize variance.

---

#### 📐 Formula

For features that take only positive values, apply:

$$
\mathbf{x' = 
\begin{cases}
\log(x),   & x > 0, \\
\log(x + 1), & x \ge 0 \;(\text{to include zeros})
\end{cases}}
$$

- **Standard log**: $x' = \log(x)$ (use only when **all** $x > 0$; no zeros or negatives).  
- **Log1p**: $x' = \log(1 + x)$ (use when $x \ge 0$ may occur, since $\log(1+0)=0$).


---

#### 🔎 When to Use

- **Skewness**: moderate right skew ($0.7 \le γ \le $1.5).  
- **Domain**: all values \(x > 0\) (use **log(1 + x)** if zeros are present).  
- **Effect**:  
  - Strongly compresses large values (tail “shrinks”).  
  - Narrows variance, making features more homoscedastic.  
- **Common in**:  
  - Financial (income, price)  
  - Count data with heavy tail (e.g. web hits)  
  - Any measurement that spans orders of magnitude

---

#### 🔄 Effect on Distribution

- **Before**: long right tail, high variance  
- **After**: closer to symmetric, smaller variance  

```python
# Q-Q Plot
import scipy.stats as stat
import pylab

def qq_plot(df, feature, title, alignment='right'):
    plt.figure(figsize=(11, 4.4))
    plt.subplot(1,2,1)
    plt.hist(df[feature], bins=30, color='green', edgecolor='black')
    plt.title("Histogram Plot", fontsize=13)
    plt.xlabel(feature, fontsize=11)
    plt.ylabel('Frequency', fontsize=11)
    skewness_value = skew(df[feature])
    x_pos = 0.05 if alignment == 'left' else 0.95
    plt.text(x=x_pos, y=0.95, s=f"Skewness: {skewness_value:.3f}", transform=plt.gca().transAxes, fontsize=12, color='darkred',
             horizontalalignment=alignment, verticalalignment='top', bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.3'))
    plt.subplot(1,2,2)
    stat.probplot(df[feature], dist='norm', plot=pylab)
    plt.suptitle(title, fontsize=15, fontweight='bold')
    plt.show()
```

```python
# Before using Log Transformation on "LSTAT" feature 
qq_plot(df, 'LSTAT', title="Before Log transformation")    # Using "LSTAT" as it is moderately Right Skewed (γ = 0.90)
df_copy = df.copy()

# Log Transformation
df_copy.loc[:, "LSTAT_log"] = np.log(df_copy["LSTAT"] + 3.3)    # Fine Tune to decrease skewness
qq_plot(df_copy, 'LSTAT_log', title="After Log transformation")
```

### Reciprocal Transformation

**Goal:** Flip and compress very large positive values to dramatically reduce strong right‑skew.

---

#### 📐 Formula

For features that take only strictly positive values, apply:

$$
\mathbf{x' = \frac{1}{x}}
$$

- **Reciprocal**: ${x' = \frac{1}{x}}$ (use only when **all** \(x > 0\); cannot include zeros or negatives).

---

#### 🔎 When to Use

- **Skewness**: very heavy right skew (γ > 1.5).  
- **Domain**: all values \(x > 0\) (no zeros or negatives).  
- **Effect**:  
  - Flips the axis so that originally large values become very small.  
  - Compresses large values most aggressively, pulling in extreme outliers.  
- **Common in**:  
  - Rates and ratios (e.g., error rates, speed = distance/time)  
  - Time‑to‑event measurements (when long durations create heavy tail)

---

#### 🔄 Effect on Distribution

- **Before**: extreme right tail with very large values  
- **After**: right tail is compressed, and former large values appear near zero  


```python
# Before using Reciprocal Transformation on "DIS" feature 
qq_plot(df, 'DIS', title="Before Reciprocal transformation")    # Using "DIS" as it is Strongly Right Skewed (γ = 1.009)

# Reciprocal Transformation
df_copy.loc[:, "DIS_repl"] = 1 / (df_copy["DIS"] + 2.2)    # Fine Tune to decrease skewness
qq_plot(df_copy, 'DIS_repl', title="After Reciprocal transformation")
```

### Exponential / Power Transformation

**Goal:**  Stretch small or negative values and accentuate larger values to reduce **left**‑skew.

---

#### 📐 Transformation Formula

For left‑skewed data, apply either a power transform or the true exponential:

$$
\mathbf{x' = 
\begin{cases}
x^k, & k > 1 \quad(\text{e.g. }k=2\text{ gives }x^2),\\[6pt]
e^{x}, & \text{(the natural exponential).}
\end{cases}}
$$

- **Power** ($x^k$): choose $k>1$ (e.g.\ $k=2$) to gently stretch the right.  
- **Exponential** ($e^{x}$): steeply magnifies differences among larger $x$.

---

#### 🔎 When to Use

- **Skewness**: moderate to strong **left** skew (γ < –0.5).  
- **Domain**: any real $x$ (power requires $x\ge0$ if $k$ is non-integer; $e^{x}$ works for all $x$).  
- **Effect**:  
  - Pulls out the left tail, making the bulk more symmetric.  
  - Amplifies larger values, increasing variance on the right.  
- **Common in**:  
  - Ratings or scores bunched at low end  
  - Features with a ceiling effect (many small values)  
  - Any measurement where high values are underrepresented

---

#### 🔄 Effect on Distribution

- **Before**: long **left** tail, compressed right.  
- **After**: more balanced; higher values more spread out.

```python
# Before using Exponential Transformation on "PTRATIO" feature 
qq_plot(df, 'PTRATIO', title="Before Exponential transformation", alignment='left')    

# Exponential Transformation
df_copy.loc[:, "PTRATIO_exp"] = np.exp(df_copy["PTRATIO"] / 3.7)    # Fine Tune to decrease skewness
qq_plot(df_copy, 'PTRATIO_exp', title="After Exponential transformation", alignment='left')
```

### Box‑Cox Transformation

**Goal:**  Find an optimal power‐law transformation to make a strictly positive feature as close to Gaussian as possible by estimating a parameter $\lambda\$.

---

#### 📐 Formula

For \(x > 0\), define

$$
\mathbf{x' = 
\begin{cases}
\displaystyle \frac{x^{\lambda} - 1}{\lambda}, & \lambda \neq 0, \\[8pt]
\log(x), & \lambda = 0.
\end{cases}}
$$

- **Box‑Cox**: $x' = \frac{x^\lambda - 1}{\lambda}$ (requires all \(x>0\)).
- **Special case**: When ($\lambda = 0$), the transform reduces to $log(x)$.  
- **Λ estimation**: $\lambda$ is chosen (e.g. via maximum likelihood) to best normalize the data.

---

#### 🔎 When to Use

- **Skewness**: moderate to strong right skew $(\gamma > 0.7$).  
- **Domain**: **strictly** \(x > 0\) (no zeros or negatives).  
- **Effect**:  
  - Warps data toward symmetry by compressing high values and expanding low values.  
  - Stabilizes variance across the range of \(x\).  
- **Common in**:  
  - Biological measures (e.g., enzyme concentrations)  
  - Economic indicators (e.g., expenditures)  
  - Any positive measurement with unknown optimal power-law behavior

---

#### 🔄 Effect on Distribution

- **Before**: long right tail, heteroscedasticity.  
- **After**: closer to bell‑shaped, more constant variance, facilitating downstream linear or Gaussian-based methods.


```python
from sklearn.preprocessing import PowerTransformer

# Before using Box-Cox  Transformation on "PTRATIO" feature 
qq_plot(df, 'PTRATIO', title="Before BOX-COX transformation", alignment='left')    

# Box-Cox Transformation
pt_box = PowerTransformer(method='box-cox', standardize=True)
df_copy.loc[:, 'PTRATIO_boxcox'] = pt_box.fit_transform(df_copy[['PTRATIO']])  # You should fit_transform the train data instead of the whole Dataframe
qq_plot(df_copy, 'PTRATIO_boxcox', title="After BOX-COX transformation", alignment='left') 
print("Parameter : ", pt_box.lambdas_) 
```

### Yeo-Johnson Transformation

**Goal:**  Handle moderate-to-heavy skewness for features that may include zeros or negative values, mapping them toward a Gaussian distribution.

---

#### 📐 Formula

$$
\mathbf{x' =
\begin{cases}
\dfrac{(x + 1)^\lambda - 1}{\lambda}, & x \ge 0,\ \lambda \neq 0,\\[6pt]
\ln(x + 1),                             & x \ge 0,\ \lambda = 0,\\[6pt]
-\dfrac{(-x + 1)^{2 - \lambda} - 1}{2 - \lambda}, & x < 0,\ \lambda \neq 2,\\[6pt]
-\,\ln(-x + 1),                        & x < 0,\ \lambda = 2.
\end{cases}}
$$

- **Parameter**: λ (learned from data to best approximate normality).  
- **Supports**: $x \ge 0$ and $x < 0$ seamlessly.  

---

#### 🔎 When to Use

- **Skewness**: moderate to strong (either left or right).  
- **Domain**: any real-valued feature (zeros and negatives allowed).  
- **Effect**:  
  - Transforms both tails toward symmetry.  
  - Stabilizes variance across feature range.  
- **Common in**:  
  - Financial returns (which can be negative or positive).  
  - Sensor measurements around a baseline.  
  - Any dataset where Box‑Cox fails due to non‑positive values.

---

#### 🔄 Effect on Distribution

- **Before**: skewed (positive or negative), unequal variance.  
- **After**: closer to symmetric bell shape, more homoscedastic.

---


```python
from sklearn.preprocessing import PowerTransformer

# Before using Yeo-Johnson Transformation on "PTRATIO" feature (As in my dataset, I have no negative feature for Transformation)
qq_plot(df, 'PTRATIO', title="Before Yeo-Johnson transformation", alignment='left')    

# Yeo-Johnson Transformation
pt = PowerTransformer(method='yeo-johnson', standardize=True)
df_copy.loc[:, 'PTRATIO_yeojohnson'] = pt.fit_transform(df_copy[['PTRATIO']])
qq_plot(df_copy, 'PTRATIO_yeojohnson', title="After Yeo-Johnson transformation", alignment='left') 
print("Parameter : ", pt.lambdas_)
```

### Quantile Transformation

**Goal:**  Map the feature distribution to a desired target distribution (e.g. uniform or Gaussian) by using the empirical quantiles, making the transformed feature distribution free of distributional assumptions and robust to outliers.

---

#### 📐 Formula

Given a feature vector **x** and its empirical cumulative distribution function $ \hat F $, the quantile-transformed value $ x' $ is

$$
x' \;=\; Q_{\text{target}}\bigl(\hat F(x)\bigr)
$$

where  
- $\hat F(x) $ is the empirical CDF of **x**, i.e. the fraction of training samples $\le x$.  
- $ Q_{\text{target}} $ is the inverse CDF (quantile function) of the target distribution:
  - **Uniform** on $[0,1]$: $Q_{\text{uni}}(u) = u$.  
  - **Gaussian** $\mathcal{N}(0,1)$: $Q_{\mathcal{N}}(u) = \Phi^{-1}(u)$, where $\Phi^{-1}$ is the standard normal inverse CDF.

---

#### 🔎 When to Use

- **Outlier robustness:** you want to dampen the influence of extreme values without explicit clipping.  
- **Non‑parametric scaling:** no assumption of linearity or normality in the original data.  
- **Heterogeneous distributions:** features with very different shapes or heavy tails.  
- **Preprocessing for algorithms:** many models benefit from inputs that are uniformly or normally distributed (e.g., kernel methods, neural networks).

---

#### 🔄 Effect on Distribution

- **Before:** arbitrary shape (skewed, multimodal, heavy‑tailed).  
- **After (Uniform):** flat distribution on \([0,1]\), equal mass in each interval.  
- **After (Gaussian):** approximately standard normal, with mean 0 and variance 1.  
- **Overall:**  
  - Spreads out dense regions and compresses sparse regions of the original distribution.  
  - Makes features comparable across different scales and shapes.

---


```python
from sklearn.preprocessing import QuantileTransformer

# Before using Quantile Transformation on "AGE" feature
qq_plot(df, 'AGE', title="Before Quantile transformation", alignment='left')    

# Quantile Transformation
qt = QuantileTransformer(n_quantiles=df.shape[0], output_distribution='uniform', random_state=42)
df_copy.loc[:, 'AGE_quantile'] = qt.fit_transform(df_copy[['AGE']])
qq_plot(df_copy, 'AGE_quantile', title="After Quantile transformation", alignment='left') 
```
